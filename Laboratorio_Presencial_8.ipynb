## Actividad

### 1. Vectores para alturas y pesos

Creen dos vectores iniciales con los valores de la tabla del enunciado.

**MODIFIQUEN EL SIGUIENTE BLOQUE DE CÓDIGO PARA ALMACENAR LOS DATOS EN PULGADAS Y LIBRAS**.

```python
pulgadas = np.array([70, 63, 72, 62, 66, 70, 74, 65, 62, 67, 65, 68],)
libras = np.array([155, 150, 180, 135, 156, 168, 178, 160, 132, 145, 139, 152])
```

Operen sobre los vectores anteriores para conseguir vectores que contengan las alturas en metros y los pesos en kilogramos. Consideren que una pulgada son $2.54$ centímetros y que una libra son aproximadamente $0.45$ kilos.

**MODIFIQUEN EL SIGUIENTE BLOQUE DE CÓDIGO PARA ALMACENAR LOS DATOS EN METROS Y KILOGRAMOS**.

```python
alturas = pulgadas * 0.0254
pesos = libras * 0.453592
```

### 2. Gráfico de pesos y alturas

Utilicen los vectores anteriores para generar el gráfico de puntos pedido.

**MODIFIQUEN EL SIGUIENTE BLOQUE DE CÓDIGO PARA MOSTRAR EL GRÁFICO DE PUNTOS**.

```python
plt.show()
plt.scatter(alturas, pesos, color='blue', label='Alturas y Pesos')
plt.xlabel('Alturas')
plt.ylabel('Pesos')
plt.title('Relación entre alturas y pesos')
plt.legend()
plt.grid(True)
plt.show()
```

### 3. Cálculo de la recta de regresión

Pueden calcular los valores de $m$ y $c$ de la ecuación $y = mx + c$ como estimen conveniente, pero se les recomienda definir las siguientes variables:

- `n`: Cantidad de puntos (pueden ocupar la función `len` sobre el vector $X$ o $Y$ para obtener su tamaño).
- `sumX`: Sumatoria de todos los valores en $X$ (pueden ocupar la función `np.sum` sobre un vector de NumPy para recibir la suma de sus valores).
- `sumY`: Sumatoria de todos los valores en $Y$.
- `sumX2`: Sumatoria de todos los valores en $X^2$.
- `sumXY`: Sumatoria de todos los valores en $XY$.
- `d`: Denominador de las fórmulas para $m$ y $c$. Su valor es `n * sumX2 - (sumX ** 2)`.

**MODIFIQUEN EL SIGUIENTE BLOQUE DE CÓDIGO PARA CALCULAR CORRECTAMENTE m Y c**.

```python
n =len(alturas)
sumX = np.sum(alturas)
sumY = np.sum(pesos)
sumX2 = np.sum(alturas**2)
sumXY = np.sum(alturas * pesos)
d = n * sumX2 - (sumX ** 2)




m = (n * sumXY - sumX * sumY) / d
c = (sumY * sumX2 - sumX * sumXY) / d
```

### 4. Gráficos finales

Hagan uso de las variables definidas en puntos anteriores para generar los gráficos pedidos. Para los extremos de la recta, utilicen puntos con coordenadas $1.5$ y $1.9$ en el eje horizontal.

**MODIFIQUEN EL SIGUIENTE BLOQUE DE CÓDIGO PARA MOSTRAR NUEVAMENTE EL GRÁFICO DE PUNTOS Y LA RECTA DE REGRESIÓN**.

```python
plt.show()
plt.scatter(alturas, pesos, color='blue', label='Alturas y Pesos')
plt.plot(alturas, m * alturas + c, color='red', label='Recta de Regresión')
plt.xlabel('Alturas (m)')
plt.ylabel('Pesos (kg)')
plt.title('Relación entre alturas y pesos')
plt.legend()
plt.grid(True)
```
